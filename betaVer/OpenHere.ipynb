### How to pair up interesting measurements for plotting
##### Using Python dictionaries - key:value pairs (x axis: y axes)

```python
# Independent variable
x_axis = "V+ [V]" 

# Dependent variables
y_axes = ["P1 [uC/cm2]", "P2 [uC/cm2]"]

# Pair them up
plots = {
    x_axis: y_axes
    # you can pair up another x axis here; then a ":"; then y_axes
    }

# Show contents
plots
```

#### Here I am using a for loop to access all independent variables (and the dependent variables we want to plot alongside them)
In this case, there is only one x_axis in the `plots` dictionary  
It will plot from two files, `BSFO13_RS800_10-150kVcm_RT.dat` and `../data\\BSFO13_RS820_10-150kVcm_RT.dat`

```python
for x_axis in plots:
    plotData(loadedTables["../data\\BSFO13_RS800_10-150kVcm_RT.dat"], x_axis, plots[x_axis])
    plotData(loadedTables["../data\\BSFO13_RS820_10-150kVcm_RT.dat"], x_axis, plots[x_axis])
```
