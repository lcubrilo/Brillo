### This code cell helps you understand what this script sees in its current folder and subfolders
We are going to be using ***relative paths***.  
They allow for more flexibility and ease of use of the notebook between different devices.  
  
The opposite of those are ***absolute paths*** such as "C:/Users/Uporabnik/Documents" which define the hard drive/partition etc.  
Relative paths ignore that, and just look from the perspective of this file.
  
The cell below tells you which files are located in the same folder as this notebook.  
Then goes to subfolders and their files, then subsubfolders and their files etc.

```python
import os
for root, dirs, files in os.walk(".", topdown=True):
    print("Files: ")
    for name in files:
        print("\t", os.path.join(root, name))
    print("Dirs: ")
    for name in dirs:
        print("\t", os.path.join(root, name))
```

What if you want to see a superfolder, not a subfolder?  
### You can backout one or more levels above, to see more subfolders
You move one folder above by adding `../` to your ***relative path***.  
You could also chain this, such as `../../../` to back out three levels of folders.  

Just play around with the value of `howManyFoldersAbove` variable, and see what the cell outputs.

```python
howManyFoldersAbove = 1 # 0
relativePath = "../" * howManyFoldersAbove
relativePath
```

Now you can define the rest of the relative path - which specific subfolders you want to go to from the previous cell (after you backed out).  
Playing with these two cells allows you to easily navigate through your folders on your PC and find the ***relative path*** of your data.

```python
nameOfSubfolders = "data" #""
relativePath += nameOfSubfolders
relativePath
```

So... which types of files are we looking for at this location?
### Here is how to narrow your query to just specific file extensions that interest you (from the path defined above)

```python
extension1 = ".dat"
extension2 = ".txt"
import os

print("The following files match the query: ")
for file in os.listdir(relativePath):
    #print("Found: {}".format(file))
    if file.endswith(extension1) or file.endswith(extension2): #This line checks which files match the condition of being ".dat" or ".txt"
        print("\t", os.path.join(relativePath, file))
```

### Setting up interesting measurements for plotting
Using Python dictionaries - key:value pairs (x axis: y axes)

```python
# Independent variable
x_axis = "V+ [V]" 

# Dependent variables
y_axes = ["P1 [uC/cm2]", "P2 [uC/cm2]"]

# Pair them up
plots = {
    x_axis: y_axes
    # you can pair up another x axis here; then a ":"; then y_axes
    }

# Let's see what we have stored in memory
plots
```
